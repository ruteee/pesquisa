```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
dist8_proc = pd.read_csv("resultados_defesa/dist8_proc.csv", index_col = 'TOUT')
dist8_proc_sv = dist8_proc[['XMEAS' + var for var in ['05','06','07','08', '09', '12', '20', '21']]]

cols_rename = ['x' + str(var)   for var in np.arange(1,9)]
dict_rename = { key:val for key, val in zip(dist8_proc_sv.columns, cols_rename)}
dist8_proc_sv = dist8_proc_sv.rename(dict_rename, axis='columns')
```

```python
dist8_proc_sv.index = pd.to_datetime(dist8_proc_sv.index, unit='s')
```
