```python
import pandas as pd
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df_te = pd.read_csv("resultados_defesa/df_te_i3eA.csv", index_col=0)
```

```python
#Plot settings

plt.rcParams['figure.dpi'] = 600
plt.rcParams['savefig.dpi'] = 600

plt.rcParams['lines.linewidth'] = 0.6
plt.rcParams['axes.linewidth'] = 0.3
plt.rcParams['font.size'] = 5

plt.rcParams['xtick.major.width'] = 0.5
plt.rcParams['ytick.major.width'] = 0.5
plt.rcParams['legend.borderpad'] = 0.2
plt.rcParams['legend.frameon']= True
plt.rcParams['legend.borderpad']= 0.1
plt.rcParams['figure.figsize']= [3,2]

plt.rcParams.update(plt.rcParamsDefault)
```

```python
# plt.rcParams
```

```python
te_vals = df_te.values.flatten()


blind_thresh = np.mean(te_vals) + 3*np.std(te_vals)

hist = plt.hist(te_vals, color= 'tab:blue' , bins=20)
plt.xlabel('Transfer Entropy')
plt.vlines(x = [0.007668474476869511, blind_thresh] , 
           ymin = [0], 
           ymax= [len(te_vals)], 
           linestyles='dashed', 
           colors = ['black', 'black'])

plt.text(x = 0.007668474476869511 + 0.003, y = 200, s = 'Proposed Threshold')
plt.text(x = blind_thresh + 0.003, y = 200, s = 'Blind Threshold')

plt.gcf().savefig("hist_te.pdf", bbox_inches='tight')


plt.show()

```

```python
sns.reset_orig()
# sns.set(style='white')

plt.rcParams['figure.dpi'] = 600
plt.rcParams['font.size'] = 5
plt.rcParams['axes.labelpad'] = 4
plt.rcParams['axes.xmargin']= 0.05
plt.rcParams['figure.subplot.wspace']= 0.5
plt.rcParams['axes.labelsize'] = 'large'


fig = plt.figure(figsize = (7,5))
ax = fig.gca()
plt.tight_layout()

axes = df_te.hist(grid=False, ax=ax)
for i,ax_set in enumerate(axes):
    for ax in ax_set:
        ax.set_xlim(0)
        ax.set_xlabel('Entropy')
        ax.xaxis.set_major_locator(plt.MaxNLocator(3))

        plt.gcf().subplots_adjust(hspace=.7)


plt.gcf().savefig("hist_vars.pdf", bbox_inches='tight')
```
